## 2.1 Data collection

In this step, a list is created to store the tickers, then setting a start and end period for the dates needed, the DataReader gets the data from Yahoo Finance using the list with the start and end date

```python
# making a list of stocks
stock_list = ['AAPL', 'J' ,'AMZN' ,'BA' ,'FB' ,'GE' ,'GOOG' ,'HD' ,'TSLA', 'JPM', 'MSFT' ,'NFLX' ,'OPK', 'PG', 'UNH', 'V']

# Setting up end and start times for data grab
#end = datetime.now()#today's date
end = datetime.datetime(2020, 7, 2)

#getting the stock for the past 10 year
start = datetime.datetime(2012, 8, 3)
#start = datetime(end.year - 8,end.month,end.day)


#For the loop to grabing yahoo finance data and then set as a dataframe
for stock in stock_list:   
    # Set dataframe as the stock ticker,globals()makes it to a globel variable
    globals()[stock] = DataReader(stock,'yahoo',start,end)
```

### Merging the CSV file

```python
import glob
import os
```

```python
# the path to the csv file directory
mycsvdir =  'C:/Users/sonia/Desktop/MScFinalProject/data' 
```

```python
# getting all the csv files in the directory with the extension .csv)
csvfiles = glob.glob(os.path.join( 'C:/Users/sonia/Desktop/MScFinalProject/data' , '*.csv'))
```

```python
# looping through the files and read them with pandas
stock_list = []  # a list to hold all the individual pandas DataFrames
for csvfile in csvfiles:
    df = pd.read_csv(csvfile)
    stock_list.append(df)
```

```python
# concatenate them all together
new_df = pd.concat(stock_list)
```
